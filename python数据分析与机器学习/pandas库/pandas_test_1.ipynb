```python
# 显示该列名下的数据

food_col = food_info['NDB_No']
```

```python
print(food_col)
```

```python
# 显示多条列名下的数据

col = ['NDB_No','Shrt_Desc']
print(food_info[col])
```

```python
# 显示已(g)结尾的数据

col_names = food_info.columns.tolist()
print(col_names)
gram_col = []

for i in col_names:
    if i.endswith('(g)'):
        gram_col.append(i)
gram_df = food_info[gram_col]
print(gram_df)
```

```python
# 将其中一列的数据除以1000后，得到的数据添加到新的列上

print(food_info['Iron_(mg)'])
Iron_1000 = food_info['Iron_(mg)']/1000
food_info.shape
food_info['Iron_(g)'] = Iron_1000
food_info.shape
```

```python
# 如果维数相同，数据之间可以进行+ - * /……

water_energy = food_info['Water_(g)']*food_info['Energ_Kcal']
print(water_energy)
```

```python
# 求某列的最值，均值……

max_cal = food_info['Energ_Kcal'].max()
print(max_cal)
```

```python
# 对以列名为xx的数据进行排序，默认sh

food_info.sort_values('Sodium_(mg)', inplace=True)
print(food_info['Sodium_(mg)'])
food_info.sort_values('Sodium_(mg)', inplace=True, ascending=False)
print(food_info['Sodium_(mg)'])
```
