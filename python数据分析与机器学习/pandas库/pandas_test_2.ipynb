```python
import numpy as np
import pandas as pd

titanic_survival = pd.read_csv('titanic_train.csv')
titanic_survival.head()
```

```python
age = titanic_survival['Age']
print(age.loc[0:10])
```

```python
age_null = pd.isnull(age)
print(age_null)
print('-------------')
age_null_t = age[age_null]
print(age_null_t)
print('-------------')
age_null_len = len(age_null_t)
print(age_null_len)
```

```python
mean_age = sum(titanic_survival['Age'])/len(titanic_survival['Age'])
print(mean_age)
```

```python
good_age = titanic_survival["Age"][age_null == False]

new_mean_age = sum(good_age)/len(good_age)
print(new_mean_age)
```

```python
corr_mean_age = titanic_survival["Age"].mean()
print(corr_mean_age)
```

```python
passenger_classes = [1,2,3]
fares_by_class = {}
for this_class in passenger_classes:
    pclass_rows = titanic_survival[titanic_survival["Pclass"]==this_class]
    pclass_fares = pclass_rows['Fare']
    fare_for_class = pclass_fares.mean()
    fares_by_class[this_class] = fare_for_class

print(fares_by_class)
```

```python
passenger_survival = titanic_survival.pivot_table(index='Pclass', values='Survived', aggfunc=np.mean)
print(passenger_survival)
```

```python
port_stats = titanic_survival.pivot_table(index='Embarked', values=['Fare','Survived'])
print(port_stats)
```

```python
drop_na_columns = titanic_survival.dropna(axis=1)
new_titanic_sur = titanic_survival.dropna(axis=0,subset=["Age","Sex"])
print(drop_na_columns)
print(new_titanic_sur)
```

```python
row_index_83_age = titanic_survival.loc[83, "Age"]
print(row_index_83_age)
```

```python
new_titanic_surv = titanic_survival.sort_values('Age', ascending=False)
print(new_titanic_surv[0:10])
print('-------------')
tit_reindexed = new_titanic_surv.reset_index(drop=True)
print(tit_reindexed)
```
