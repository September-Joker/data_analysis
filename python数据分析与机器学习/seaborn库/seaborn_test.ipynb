```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
```

```python
print(help(sns))
```

```python
def sinplot(filp=1):
    x=np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*filp)

sinplot()
```

```python
sns.set()
sinplot()
```

```python
sns.set_style('whitegrid')
# sns.set_style('darkgrid')
# sns.set_style('white')
# sns.set_style('dark')
# sns.set_style('ticks')
data = np.random.normal(size=(20,6))+np.arange(6)/2
sns.boxplot(data=data)
```

```python
sns.set_style('ticks')
sinplot()
# 去掉右边与上边的轴
sns.despine()
```

```python
sns.violinplot(data)
sns.despine(offset=10)
```

```python
sns.set_style('whitegrid')
sns.boxplot(data=data, palette='deep')
sns.despine(left=True,bottom=True)
# print(help(sns.despine))
```

```python
with sns.axes_style('darkgrid'):
    plt.subplot(2,1,1)
    # 逗号可以省略，亦可以写作plt.subplot(211)
    sinplot()
plt.subplot(2,1,2)
sinplot(-1)
```
