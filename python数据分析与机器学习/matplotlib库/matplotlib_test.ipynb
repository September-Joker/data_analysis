```python
data['MONTH'] = data['DATE'].dt.month
data['MONTH'] = data['DATE'].dt.month

fig = plt.figure()

plt.plot(data[0:12]['MONTH'],data[0:12]['VALUE'],c='r')
plt.plot(data[12:24]['MONTH'],data[12:24]['VALUE'],c='b')

plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
colors = ['red','blue','green','orange','black']
for i in range(5):
    start_index =i*12
    end_index = (i+1)*12
    subset = data[start_index:end_index]
    label = str(1948+i)
    plt.plot(subset["MONTH"],subset["VALUE"],c=colors[i],label= label)
plt.legend(loc='best')
plt.xlabel('month')
plt.ylabel('unployment rate')
plt.show()
```

```python
import pandas as pd

film = pd.read_csv('fandango_scores.csv')
cols = ['FILM','RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue','Fandango_Stars']

col = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue','Fandango_Stars']

norm_reviews = film[cols]

bar_hight = norm_reviews.ix[0,col].values
# 柱高
print(bar_hight)

bar_positions = np.arange(5) + 0.75
# 柱子离原点的距离
print(bar_positions)

fig, ax = plt.subplots()

ax.bar(bar_positions,bar_hight,0.3)# 0.3表示柱子的宽度
ax.set_ylabel('rating source')
ax.set_xlabel('average rating')
ax.set_title('Average User Rating For Averagers: Age of Ultron(2015)')

plt.show()
```

```python
num_col = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue','Fandango_Stars']

bar_widths = norm_reviews.ix[0,num_col].values
bar_positions = np.arange(5)+0.75

tick_positions = range(1,6)
fix,ax = plt.subplots()

# barh表示柱状图为横着的
# bar 表示柱状图是竖着的
ax.barh(bar_positions, bar_widths, 0.5)

ax.set_yticks(tick_positions)
ax.set_yticklabels(num_col)

ax.set_xlabel('rating source')
ax.set_ylabel('average rating')
ax.set_title('Average User Rating For Averagers: Age of Ultron(2015)')
plt.show()
```
